```python
import pandas as pd
import matplotlib.pyplot as plt
from numpy import arange

reviews = pd.read_csv("fandango_score_comparison.csv")
cols = ["FILM", "RT_user_norm", "Metacritic_user_nom", "IMDB_norm", "Fandango_Ratingvalue","Fandango_Stars"]
norm_reviews = reviews[cols]
num_cols = ["RT_user_norm", "Metacritic_user_nom", "IMDB_norm", "Fandango_Ratingvalue","Fandango_Stars"]

bar_heights = norm_reviews.ix[0, num_cols].values  #指当前条形柱的高度
print(bar_heights)
bar_position = arange(5)+ 1  #指每个柱离原点多远
print(bar_position)
fig, ax = plt.subplots()  #创建图形
ax.bar(bar_position, bar_heights, 0.3)  #画出条形图，0.3指每个柱的宽度
tick_position = range(1,6)

ax.set_xticks(tick_position)
ax.set_xticklabels(num_cols, rotation = 45)
ax.set_xlabel("Rating Source")
ax.set_ylabel("Average Rating")
ax.set_title("Average User Rating For Averagers: Age of Ultron (2015)")
plt.show()
```
