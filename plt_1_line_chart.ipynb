```python
import pandas as pd
unrate = pd.read_csv("UNRATE.csv")
unrate["DATE"] = pd.to_datetime(unrate["DATE"]) #将1948/1/1这种时间格式变成1948-01-01，这样看着方便
print(unrate.head(12))
```

```python
import matplotlib.pyplot as plt
plt.plot()
plt.show()
```

```python
first_twelve = unrate[0:12]
plt.plot(first_twelve["DATE"], first_twelve["VALUE"])  #指定X，Y轴
plt.xticks(rotation = 45) #指让横坐标标识变成45度表示
plt.xlabel("Month")
plt.ylabel("Unemployment Rate")
plt.title("Monthly Unemployment Trends, 1948")
plt.show()
```

```python
fig = plt.figure()  #先指定出画图的区间
ax1 = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax4 = fig.add_subplot(2,2,4) #没指定2,2,3，第三个图就不会画出来
plt.show()
```

```python
import numpy as np
fig = plt.figure(figsize = (3, 3)) #指定了整个画图区域的大小
ax1 = fig.add_subplot(2,1,1)
ax2 = fig.add_subplot(2,1,2)

ax1.plot(np.random.randint(1, 5, 5), np.arange(5)) #分别给两个子图画图
ax2.plot(np.arange(10)*3, np.arange(10))
plt.show()
```

```python
unrate["MONTH"] = unrate["DATE"].dt.month  #只取日期中的月，给unrate加了一列
print(unrate[0:5])
fig = plt.figure(figsize = (6, 3))
plt.plot(unrate[0:12]["MONTH"], unrate[0:12]["VALUE"], c="red")
plt.plot(unrate[12:24]["MONTH"], unrate[12:24]["VALUE"], c="blue") #一个图中画两条线，不同颜色
plt.show()
```

```python
#查看前5年的图像
fig = plt.figure(figsize = (10, 6))
colors = ["red", "blue", "black", "yellow", "pink"]
for i in range(5):
    start_index = i*12
    end_index = (i+1)*12
    label = str(1948+i)  #label就是指表示的图例
    plt.plot(unrate["MONTH"][start_index: end_index], unrate["VALUE"][start_index: end_index], c= colors[i], label = label)
plt.legend(loc= "best")  #只写label是不行的，只有通过.legend才会显示出来，best指把图例框自定义放在最佳位置，还有别的可选，用print(help(plt.legend))查看
plt.show()
```
